### 2. Imagine you are given a list of items in a DataFrame as below.D=["A","B","C","D","E","AA","AB"].Now convert and display this categorical data using dummies function of pandas and label encoder function of Scikit-learn.


```python
df = pd.DataFrame({'D': ["A","B","C","D","E","AA","AB"],
'B': [10,20,30,40,50,60,70]}, index=[0,1,2,3,4,5,6])
print(df)
df_dummies= pd.get_dummies(df, prefix='A', columns=['D'])
df_dummies
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['A_AB'] = le.fit_transform(df.D)
print(df)
```

### 3. Read a csv details using pandas dataframe then apply min-max normalization and z-score normalization over the each column of the dataset

```python
from sklearn import datasets
from sklearn import preprocessing
iris_ = datasets.load_iris()
X = iris_.data
y = iris_.target
print(X)
minmax_scale = preprocessing.MinMaxScaler().fit(X)
X_minmax = minmax_scale.transform(X)
print(X_minmax)
```

```python
std_scale = preprocessing.StandardScaler().fit(X)
X_std = std_scale.transform(X)
print(X_std)
```
